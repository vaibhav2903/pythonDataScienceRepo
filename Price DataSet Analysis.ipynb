```python
import scipy.stats as stats
import math
```

```python
np.random.seed(10)

sample_size=1000
sample=np.random.choice(a=df['SalePrice'],size=sample_size)
sample_mean = sample.mean()

z_critical = stats.norm.ppf(q=0.95)
print(z_critical)
pop_stdev = df['SalePrice'].std()

margin_of_Error= z_critical*(pop_stdev/math.sqrt(sample_size))
confidence_interval = (sample_mean - margin_of_Error,sample_mean + margin_of_Error)
print(confidence_interval)
print(df['SalePrice'].mean())
```

```python
# capturing true mean with several confidence intervals

np.random.seed(12)
sample_size = 500

intervals =[]
sample_means = []

for sample in range(100):
    sample = np.random.choice(a=df['SalePrice'],size=sample_size)
    sample_mean = sample.mean()
    sample_means.append(sample_mean)
    
    z_critical = stats.norm.ppf(q = 0.975)
    pop_stdev = df['SalePrice'].std()
    
    stats.norm.ppf(q=0.025)
    
    margin_of_error=z_critical* (pop_stdev/math.sqrt(sample_size))
    
    confidence_interval = (sample_mean - margin_of_error,sample_mean + margin_of_error)
    intervals.append(confidence_interval)
    
plt.figure(figsize=(13,9))

plt.errorbar(x=np.arange(0.1,100,1),y = sample_means,yerr=[(top-bot)/2 for top,bot in intervals], fmt='o')
plt.hlines(xmin=0,xmax=100, y = df['SalePrice'].mean(),linewidth=2.0,color="red")
#plt.hlines(xmin=,xmax=100,y = )
plt.title('Confidence intervals for 100  trials' ,fontsize=20)
plt.show()
```

```python
#comparing OldTown prices with all prices

from statsmodels.stats.weightstats import ztest

z_statistic,p_value=ztest(x1=df[df['Neighborhood']=='OldTown']['SalePrice'],value=df['SalePrice'].mean())

print('Z-statistic: {}'.format(z_statistic))
print('p-value: {:.50f}'.format(p_value))
```
